```python
import pandas as pd

# Yahoo Finance에서 제공 금융 데이터 로드
import yfinance as yf

# 2023 미국주식 시가총액 순위 Top 10 (2023.05.30 기준)
# 데이터 출처 Yahoo Finance
# 금융 데이터 로드
tickers = ['AAPL','MSFT','GOOGL','AMZN','NVDA','META','TSLA', 'PEP','AVGO','ASML']
start = '2018-01-01'
end = '2023-05-31'
df_ticker_info = pd.DataFrame()
for ticker in tickers:
    ticker_yf = yf.Ticker(ticker)
    if df_ticker_info.empty:
        df_ticker_info = ticker_yf.history(start = start, end = end)
        df_ticker_info['ticker']= ticker 
    else:
        data_temp = ticker_yf.history(start = start, end = end)
        data_temp['ticker']= ticker 
        df_ticker_info = pd.concat([df_ticker_info, data_temp])

df_ticker_info.head()
```

```python
import numpy as np

# pct_change 함수를 이용, 각 날짜별 종가를 이용하여 수익률을 계산
df_ticker_info['currRet'] = df_ticker_info['Close'].pct_change()

# ticker가 바뀔 때는 정확한 수익률을 계산할 수 없으므로 첫행은 NaN 처리
temp_df_ticker_info = df_ticker_info.loc[:, 'currRet'].copy()
for i in range(len(df_ticker_info)-1):
  if df_ticker_info.iloc[i]['ticker'] != df_ticker_info.iloc[i+1]['ticker']:
    temp_df_ticker_info.iloc[i+1] = np.nan
    
df_ticker_info['currRet'] = temp_df_ticker_info

# 전일, 당일, 후일의 데이터를 활용하여 이벤트 수익률을 계산하여 이벤트에 따른 주가의 변동을 보다 정확하게 측정
# currRet(t-1) + currRet(t) + currRet(t+1) -> 뉴스 데이터 전일, 당일, 후일의 수익
df_ticker_info['eventRet'] = df_ticker_info['currRet'] + df_ticker_info['currRet'].shift(-1) + df_ticker_info['currRet'].shift(1)

# 날짜 (년-월-일) 정리
df_ticker_info.reset_index(inplace=True)
df_ticker_info['date'] = pd.to_datetime(df_ticker_info['Date']).dt.date

df_ticker_info.head()
```

```python
# seekingalpha.com에서 뉴스 헤드라인 정보 스크래핑
import requests
from bs4 import BeautifulSoup
import json
from datetime import datetime
import pandas as pd

exit_flag = False
df = pd.DataFrame(columns=['ticker', 'headline', 'publishDate'])
i = 15987
tickers = ['AAPL','MSFT','GOOGL','AMZN','NVDA','META','TSLA', 'PEP','AVGO','ASML']

while True:
  # 첫번째 while 루프 페이지 번호 증가
  if exit_flag:
    break
    
  url = "https://seekingalpha.com/market-news"
  url = url + "?page=" + str(i)
  response = requests.get(url)
  page = requests.get(url)

  soup = BeautifulSoup(page.text, 'html.parser')
  scripts = soup.find_all('script')

  # 로드할 데이터 시간 (2018년 1월 1일까지의 데이터 로드)
  timeLimit = datetime.strptime('2018-12-31', '%Y-%m-%d')

  for j in range(0, 24):
    # 두번째 for 루프 json 내 marketNews, response, data 배열 순회

    for script in scripts:
      # 세번째 for 루프 로드된 scripts 를 라인별로 순회
      if 'window.SSR_DATA =' in script.text:
        json_data = script.text.split('window.SSR_DATA = ')[1]
        json_data = json_data[:-1]
        data = json.loads(json_data)

        headline = data['marketNews']['response']['data'][j]['attributes']['title']
        newsId = ""
        
        if len(data['marketNews']['response']['data'][j]['relationships']['primaryTickers']['data']) == 0:
          newsId = ""
        else:
          newsId = data['marketNews']['response']['data'][j]['relationships']['primaryTickers']['data'][0]['id']

        publishDate = data['marketNews']['response']['data'][j]['attributes']['publishOn']

        publishTime_naive = datetime.fromisoformat(publishDate)
        publishTime_naive = publishTime_naive.strftime('%Y-%m-%d %H:%M:%S')
        publishTime_naive = datetime.strptime(publishTime_naive, '%Y-%m-%d %H:%M:%S')

        if timeLimit > publishTime_naive:      
          exit_flag = True
          break

        if exit_flag:
          break
        count = 0
        stock = ""
        ticker = ""
        # id 기반으로 ticker 찾기
        # Top10 항목에 포함된 주식으로만 데이터 프레임을 구축
        while True:
          if newsId == "":
            break

          if len(data['marketNews']['response']['included']) == 0:
            stock = ""
          else:
            stock = data['marketNews']['response']['included'][count]['id']
            
          if newsId == stock:
            ticker = data['marketNews']['response']['included'][count]['attributes']['name']
            if ticker in tickers:
              new_row = pd.DataFrame({'ticker': [ticker], 'headline': [headline], 'publishDate': [publishTime_naive]})
              df = pd.concat([df, new_row])          
            break
          count = count + 1

      # 세번째 for 루프 종료
    
    # 두번째 for 루프 종료
  
  i = i + 1
  # 첫번째 while 루프 종료

df
```

```python
# 뉴스 헤드라인 정보를 스크래핑하는데 10시간 이상의 시간이 소요 됨
# 끊어서 헤드라인 정보를 로드하고 csv 파일(news.csv)로 만들어 병합한 뒤 google drive에 저장하여 사용
from google.colab import drive
drive.mount('/content/drive/')
import pandas as pd

df_news = pd.read_csv('/content/drive/MyDrive/DeepNeuralNetwork/news.csv')

df_news

# 날짜 (년-월-일) 정리
df_news['date'] = pd.to_datetime(df_news['publishDate']).dt.date

# df_news 에 df_ticker_info 를 left 조인 
df_mergedData = pd.merge(df_news, df_ticker_info, how='left', left_on=['date','ticker'], right_on=['date','ticker'])

# 필요한 데이터만 남기고 정리
df_finalData = df_mergedData[['ticker','headline','date','eventRet','Close']]

# NaN 이 있는 행 드롭
df_finalData = df_finalData.dropna()

# 모델 분석 시 편리하게 사용하기 위하여 전처리 된 정보 csv 파일로 저장
df_finalData.to_csv('/content/drive/MyDrive/DeepNeuralNetwork/finalData.csv')

df_finalData.head()
```

```python
# 자연어 처리, 문장 분류, 개체 인식, 감정 분석, 질문 응답 등의 작업을 처리할 수 있는 SpaCy에서 제공하는 en_core_web_lg 모델 설치
#!python -m spacy download en_core_web_lg

import pandas as pd
import en_core_web_lg
nlp = en_core_web_lg.load()

# 감정 분석에 어떤 모델을 이용할지 여러 알고리즘을 이용하여 확인
# kaggle에서 모델 확인에 필요한 감정 데이터가 레이블링 된 데이터를 로드 함 
# (https://www.kaggle.com/datasets/ryanchan911/selective-stock-headlines-sentiment?select=Project6500.csv)
# 해당 데이터를 다운 받아 구글 드라이드에 저장하여 사용
df_sentimentData = pd.read_csv('/content/drive/MyDrive/DeepNeuralNetwork/Project6500.csv', encoding='ISO-8859-1')

# 로드 된 값의 모든 headline에 대해 단어 벡터 평균을 계산
# 해당 평균 벡터는 각 문장의 의미를 임베딩(벡터 공간으로 표현)하는데 사용 됨
vectors = pd.np.array([pd.np.array([token.vector for token in nlp(s)]).mean(axis=0) * pd.np.ones((300)) \
                           for s in df_sentimentData['headline']])
```

```python
# 모델 선정을 위한 각종 라이브러리 로드
from sklearn.model_selection import train_test_split, KFold, cross_val_score, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# 데이터프레임의 "sentiment" 컬럼을 타겟 변수로 지정하여 변수 Y에 할당
Y= df_sentimentData["sentiment"]
# vectors 변수를 독립 변수로 지정하여 변수 X에 할당
X = vectors

validation_size = 0.3
seed = 7
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = validation_size, random_state = seed)

# 모델 평가를 위한 옵션 설정
num_folds = 10
seed = 7
scoring = 'accuracy'

# 분류 모델 설정
models = []
models.append(('LR', LogisticRegression()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('SVM', SVC()))
models.append(('NN', MLPClassifier()))
models.append(('RF', RandomForestClassifier()))

results = []
names = []
kfold_results = []
test_results = []
train_results = []
for name, model in models:
  # 모델 교차 검증 준비
  kfold = KFold(n_splits=num_folds, random_state = seed, shuffle=True)
  # 교차 검증을 통한 모델 평가
  cv_results = cross_val_score(model, X_train, Y_train, cv=kfold, scoring=scoring)
  results.append(cv_results)
  names.append(name)

  # 전체 학습 및 테스트를 통한 모델 평가  
  res = model.fit(X_train, Y_train)
  # 학습 데이터에 대한 예측 결과를 평가하여 학습 정확도를 계산
  train_result = accuracy_score(res.predict(X_train), Y_train)
  train_results.append(train_result)
  # 테스트 데이터에 대한 예측 결과를 평가하여 테스트 정확도를 계산
  test_result = accuracy_score(res.predict(X_test), Y_test)
  test_results.append(test_result)

  # 모델 이름, 교차 검증 평균 정확도, 교차 검증 정확도의 표준 편차, 학습 정확도, 테스트 정확도
  msg = "%s: %f (%f) %f %f" % (name, cv_results.mean(), cv_results.std(), train_result, test_result)
  print(msg)
```

```python
import numpy as np
# 바 그래프를 이용한 결과 비교
from matplotlib import pyplot

fig = pyplot.figure()
ind = np.arange(len(names))
width = 0.3
fig.suptitle('Comparison of model result')
ax = fig.add_subplot(111)
pyplot.bar(ind - width/2, train_results,  width = width, label = 'Train Results', color = 'royalblue')
pyplot.bar(ind + width/2, test_results, width = width, label = 'Test Results', color = 'skyblue')
fig.set_size_inches(10,6)
pyplot.legend()
ax.set_xticks(ind)
ax.set_xticklabels(names)
pyplot.show()
```

```python
from keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# LSTM을 이용한 감정 분석 시도
vocabulary_size = 20000
tokenizer = Tokenizer(num_words= vocabulary_size)
tokenizer.fit_on_texts(df_sentimentData['headline'])
sequences = tokenizer.texts_to_sequences(df_sentimentData['headline'])

# LSTM 모델의 입력 값 설정
X_LSTM = pad_sequences(sequences, maxlen=50)
Y_LSTM = df_sentimentData["sentiment"]

X_train_LSTM, X_test_LSTM, Y_train_LSTM, Y_test_LSTM = train_test_split(X_LSTM, Y_LSTM, test_size = validation_size, random_state = seed)

# LSTM 관련 라이브러리 추가
from keras.wrappers.scikit_learn import KerasClassifier
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense

# LSTM 모델을 생성
def create_model(input_length=50):
    model = Sequential()
    model.add(Embedding(20000, 300, input_length=50))
    model.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])    
    return model

# LSTM 모델 훈련
model_LSTM = KerasClassifier(build_fn = create_model, epochs = 3, verbose = 1, validation_split = 0.4)
model_LSTM.fit(X_train_LSTM, Y_train_LSTM)
```

```python
# LSTM 훈련 결과 평가
train_result_LSTM = accuracy_score(model_LSTM.predict(X_train_LSTM), Y_train_LSTM)
test_result_LSTM = accuracy_score(model_LSTM.predict(X_test_LSTM), Y_test_LSTM)

print(train_result_LSTM,test_result_LSTM)
```

```python
# LSTM 데이터 추가 후 그래프 출력
train_results.append(train_result_LSTM)
test_results.append(test_result_LSTM)
names.append("LSTM")

import numpy as np
# 바 그래프를 이용한 결과 비교
from matplotlib import pyplot

fig = pyplot.figure()
ind = np.arange(len(names))
width = 0.3
fig.suptitle('Comparison of model result')
ax = fig.add_subplot(111)
pyplot.bar(ind - width/2, train_results,  width = width, label = 'Train Results', color = 'royalblue')
pyplot.bar(ind + width/2, test_results, width = width, label = 'Test Results', color = 'skyblue')
fig.set_size_inches(10,6)
pyplot.legend()
ax.set_xticks(ind)
ax.set_xticklabels(names)
pyplot.show()
```
